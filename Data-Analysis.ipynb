```python
import pandas as pd
import numpy as np
```

```python
FILE_PATH = r"./data/IPL.csv"
```

```python
print(f"-"*50)
print(f"Reading IPL Data")
ipl_data = pd.read_csv(FILE_PATH, low_memory=False)
print(f"Read IPL Data with shape {ipl_data.shape}")
print(f"Columns found are \n {ipl_data.columns}")
print(f"-"*50)
```

```python
print(f"-"*50)
print(ipl_data.info())
print(f"-"*50)
```

```python
print(f"-"*50)
print(ipl_data['overs'].value_counts())
print(f"-"*50)
```

```python
print(f"-"*50)
columns_to_drop = ['Unnamed: 0', 'match_type', 'event_name', 'ball', 'runs_not_boundary', 'day', 'month', 'year', 'gender', 'team_type', 'balls_per_over', 'overs', 'over', 'event_match_no', 'match_number', 'batting_partners', 'method', 'new_batter', 'next_batter']
cleaned_ipl_data = ipl_data.drop(columns=columns_to_drop)
print(cleaned_ipl_data.info())
print(f"-"*50)
```

```python
print(f"-"*50)
# -------------- Standardize Team Name --------------
cleaned_ipl_data = cleaned_ipl_data.replace({"Rising Pune Supergiant":"Rising Pune Supergiants"})
print(f"Replaced 'Rising Pune Supergiant' with 'Rising Pune Supergiants'")


# -------------- Standardize Venue Name --------------
cleaned_ipl_data["venue"] = cleaned_ipl_data["venue"].map(lambda x: x.split(",")[0].replace("."," ").replace(" IS Bindra ", " ").strip() if isinstance(x,str) else x)
cleaned_ipl_data["venue"] = cleaned_ipl_data["venue"].replace({
    "Sardar Patel Stadium": "Narendra Modi Stadium",
    "Feroz Shah Kotla": "ARUN JAITLEY STADIUM",
    "Zayed Cricket Stadium": "SHEIKH ZAYED STADIUM"
})
print(f"Cleaned and standardized venue names")


# -------------- Standardize Season Numbers --------------  
cleaned_ipl_data["season"] = cleaned_ipl_data["season"].replace(
    {
        "2007/08":"2008",
        "2009/10":"2010",
        "2020/21":"2020"
    }
)
print(f"Cleaned and standardized season numbers")


# -------------- Standardize Match Stage --------------  
cleaned_ipl_data["stage"] = cleaned_ipl_data["stage"].replace(
    {
        "Unknown": "League"
    }
)
print(f"Cleaned and standardized match stage numbers")


# -------------- Fill Unknown City Name -------------- 
cleaned_ipl_data["city"] = np.where(
    cleaned_ipl_data["city"]=="Unknown",
    cleaned_ipl_data["venue"].str.split(" ").str[0],
    cleaned_ipl_data["city"]
)
cleaned_ipl_data["city"] = cleaned_ipl_data["city"].replace(
    {
        "Bengaluru": "Bangalore",
        "Navi Mumbai": "Mumbai",
        "New Chandigarh": "Chandigarh"
    }
)
print(f"Cleaned and standardized City Names")


print(f"-"*50)
```

```python
print(f"-"*50)
print(cleaned_ipl_data.info())
print(f"-"*50)
```

```python
print(f"-"*50)

# =================================================
#               Categorical Columns
# =================================================
categorical_columns = ["batting_team", "bowling_team", "batter", "bowler","extra_type", "non_striker", "wicket_kind", "player_out", "fielders", "review_batter", "umpire", "player_of_match", "match_won_by", "win_outcome", "toss_winner", "toss_decision", "venue", "city", "superover_winner", "result_type", "stage", "team_reviewed", "review_decision"]
for col in categorical_columns:
    try:
        cleaned_ipl_data[col] = cleaned_ipl_data[col].astype(str)
        cleaned_ipl_data[col] = cleaned_ipl_data[col].map(lambda x: x.strip().upper() if isinstance(x,str) else x)
    except Exception as e:
        raise Exception(f"Failed to convert {col} column to string.")
print(f"Identified and type-casted {len(categorical_columns)} categorical columns")


# =================================================
#               Numerical Columns
# =================================================
numerical_columns = ["match_id", "innings","bat_pos", "runs_batter", "balls_faced", "runs_extras", "runs_total", "runs_bowler", "non_striker_pos", "runs_target", "season", "team_runs", "team_balls", "team_wicket", "batter_runs", "batter_balls", "bowler_wicket", "ball_no"]
for col in numerical_columns:
    try:
        cleaned_ipl_data[col] = cleaned_ipl_data[col].astype(float)
    except Exception as e:
        raise Exception(f"Failed to convert {col} column to float.")
print(f"Identified and type-casted {len(numerical_columns)} float columns")


# =================================================
#               Boolean Columns
# =================================================
boolean_columns = ["valid_ball", "umpires_call", "striker_out"]
for col in boolean_columns:
    try:
        cleaned_ipl_data[col] = cleaned_ipl_data[col].astype(bool)
    except Exception as e:
        raise Exception(f"Failed to convert {col} column to boolean.")
print(f"Identified and type-casted {len(boolean_columns)} boolean columns")


# =================================================
#               Date Columns
# =================================================
date_columns = ["date"]
for col in date_columns:
    try:
        cleaned_ipl_data[col] = pd.to_datetime(cleaned_ipl_data[col]).dt.date
    except Exception as e:
        raise Exception(f"Failed to convert {col} column to date.")
print(f"Identified and type-casted {len(date_columns)} date columns")


# =================================================
#               Total (All) Columns
# =================================================
total_columns = len(categorical_columns) + len(numerical_columns) + len(boolean_columns) + len(date_columns)
print(f"Total {total_columns} columns identified and type-casted correctly!")

print(f"-"*50)
```

### Enriched Delivery Fact Table

```python
enriched_delivery_fact_table = delivery_fact_table.copy()
```

```python
def find_phase(row):
    if row["innings"]>2:
        return "Super Over"
    elif 1<=row["over_number"]<=6:
        return "Powerplay Over"
    elif 6<=row["over_number"]<=15:
        return "Middle Over"
    elif 16<=row["over_number"]<=20:
        return "Death Over"
    else:
        return None

# -----------------is_dot_ball-----------------
enriched_delivery_fact_table["is_dot_ball"] = np.where(
    enriched_delivery_fact_table["runs_total"]==0,
    True,
    False
)

# -----------------is_four-----------------
enriched_delivery_fact_table["is_four"] = np.where(
    enriched_delivery_fact_table["runs_batter"]==4,
    True,
    False
)

# -----------------is_six-----------------
enriched_delivery_fact_table["is_six"] = np.where(
    enriched_delivery_fact_table["runs_batter"]==6,
    True,
    False
)

# -----------------is_boundary-----------------
enriched_delivery_fact_table["is_boundary"] = np.where(
    (enriched_delivery_fact_table["is_four"] | enriched_delivery_fact_table["is_six"]),
    True,
    False
)

# -----------------is_wicket-----------------
enriched_delivery_fact_table["is_wicket"] = np.where(
    enriched_delivery_fact_table["wicket_kind"].notna(),
    True,
    False
)

# -----------------is_bowler_wicket-----------------
enriched_delivery_fact_table["bowler_wicket"] = enriched_delivery_fact_table["bowler_wicket"].replace({0:False, 1:True})
enriched_delivery_fact_table = enriched_delivery_fact_table.rename(columns={
    "bowler_wicket": "is_bowler_wicket"
})

# -----------------over_number-----------------
enriched_delivery_fact_table["over_number"] = np.floor(enriched_delivery_fact_table["ball_no"])+1

# -----------------over phase-----------------
enriched_delivery_fact_table["phase"] = enriched_delivery_fact_table.apply(find_phase, axis=1)
```

### Agg Player Batting Table

```python
player_id_col = []
season_id_col = []
matches_col = []
innings_col = []
runs_col = []
balls_col = []
average_col = []
strike_rate_col = []
fours_col = []
sixes_col = []
fifties_col = []
hundreds_col = []
dot_ball_pct_col = []
boundary_pct_col = []
highest_score_col = []


season_map = season_dim_table.set_index("season")["season_id"]
player_map = player_dim_table.set_index("player_name")["player_id"]

for key,values in cleaned_ipl_data.groupby(by=["batter", "season"]):

    # -------------------------------------------------------------
    #                       Calculate Values
    # -------------------------------------------------------------
    runs = values["runs_batter"].sum()
    matches = values["match_id"].nunique()
    balls = values["balls_faced"].sum()

    try:
        fours = values["runs_batter"].value_counts()[4]
    except:
        fours = 0
    
    try:
        sixes = values["runs_batter"].value_counts()[6]
    except:
        sixes=0
    
    fifties = 0
    hundereds = 0
    highest_score = -1
    for _,v in values.groupby(["match_id", "innings"]):
        runs_total = v["runs_batter"].sum()
        if runs_total>=100:
            hundereds += 1
        elif runs_total>=50:
            fifties += 1
        if runs_total>highest_score:
            highest_score = runs_total
    try:
        dot_balls_pct = round(values["runs_total"].value_counts()[0]/balls * 100, 2)
    except KeyError:
        dot_balls_pct = 0
    
    boundary_pct = round((fours+sixes)/balls*100, 2)

    # -------------------------------------------------------------
    #                       Append Values
    # -------------------------------------------------------------
    player_id_col.append(player_map[key[0]])
    season_id_col.append(season_map[key[1]])
    matches_col.append(matches)
    innings_col.append(values[["match_id","innings"]].drop_duplicates().shape[0])
    runs_col.append(runs)
    balls_col.append(balls)
    average_col.append(round(runs/matches, 2))
    strike_rate_col.append(round(runs/balls*100, 2))
    fours_col.append(fours)
    sixes_col.append(sixes)
    fifties_col.append(fifties)
    hundreds_col.append(hundereds)
    dot_ball_pct_col.append(dot_balls_pct)
    boundary_pct_col.append(boundary_pct)
    highest_score_col.append(highest_score)
```

```python
# -------------------------------------------------------------
#                       Build Dataframe
# -------------------------------------------------------------

agg_player_batting_table = pd.DataFrame()
agg_player_batting_table["player_id"] = player_id_col
agg_player_batting_table["season_id"] = season_id_col
agg_player_batting_table["matches"] = matches_col
agg_player_batting_table["innings"] = innings_col
agg_player_batting_table["runs"] = runs_col
agg_player_batting_table["balls"] = balls_col
agg_player_batting_table["average"] = average_col
agg_player_batting_table["strike_rate"] = strike_rate_col
agg_player_batting_table["fours"] = fours_col
agg_player_batting_table["sixes"] = sixes_col
agg_player_batting_table["fifties"] = fifties_col
agg_player_batting_table["hundreds"] = hundreds_col
agg_player_batting_table["dot_ball_pct"] = dot_ball_pct_col
agg_player_batting_table["boundary_pct"] = boundary_pct_col
agg_player_batting_table["highest_score"] = highest_score_col

agg_player_batting_table.head()
```

```python
agg_player_batting_table.to_csv("database-tables/agg-player-batting-table.csv",index=False)
```

### Agg Player Bowling Table

```python
player_id_col = []
season_id_col = []
matches_col = []
innings_col = []
balls_bowled_col = []
overs_col = []
runs_conceded_col = []
wickets_col = []
economy_col = []
bowling_avg_col = []
bowling_sr_col = []
dot_balls_col = []
dot_balls_pct_col = []
fours_conceded_col = []
sixes_conceded_col = []
best_bowling_figs_col = []

season_map = season_dim_table.set_index("season")["season_id"]
player_map = player_dim_table.set_index("player_name")["player_id"]

for key,values in cleaned_ipl_data.groupby(by=["bowler", "season"]):

    # -------------------------------------------------------------
    #                       Calculate Values
    # -------------------------------------------------------------
    runs = values["runs_batter"].sum()
    matches = values["match_id"].nunique()
    balls = values["valid_ball"].value_counts()[True]
    values["over"] = np.ceil(values["ball_no"])
    overs = values[["match_id","over"]].drop_duplicates().shape[0]
    runs = values["runs_bowler"].sum()
    wickets = values["bowler_wicket"].sum()
    economy = round(runs/overs, 2)
    bowling_avg = round(runs/wickets, 2) if wickets!=0 else 0
    bowling_sr = round(balls/wickets, 2) if wickets!=0 else 0
    try:
        dot_balls = values["runs_total"].value_counts()[0]
        dot_balls_pct = round(dot_balls/balls * 100, 2)
    except KeyError:
        dot_balls = 0
        dot_balls_pct = 0

    try:
        fours = values["runs_batter"].value_counts()[4]
    except:
        fours = 0
    
    try:
        sixes = values["runs_batter"].value_counts()[6]
    except:
        sixes = 0
    
    most_wicks = -1
    most_runs = -1
    for k,v in values.groupby(["match_id"]):

        if (v["bowler_wicket"].sum()>most_wicks) or (v["bowler_wicket"].sum()==most_wicks and v["runs_bowler"].sum()<most_runs):
            most_wicks = int(v["bowler_wicket"].sum())
            most_runs = int(v["runs_bowler"].sum())

    bbf = f"{most_wicks}/{most_runs}"

    # -------------------------------------------------------------
    #                       Append Values
    # -------------------------------------------------------------
    player_id_col.append(player_map[key[0]])
    season_id_col.append(season_map[key[1]])
    matches_col.append(matches)
    innings_col.append(values[["match_id","innings"]].drop_duplicates().shape[0])
    balls_bowled_col.append(balls)
    overs_col.append(overs)
    runs_conceded_col.append(runs)
    wickets_col.append(wickets)
    economy_col.append(economy)
    bowling_avg_col.append(bowling_avg)
    bowling_sr_col.append(bowling_sr)
    dot_balls_col.append(dot_balls)
    dot_balls_pct_col.append(dot_balls_pct)
    fours_conceded_col.append(fours)
    sixes_conceded_col.append(sixes)
    best_bowling_figs_col.append(bbf)

    if key[0]=="B KUMAR":
        print(key[0], key[1], matches, balls, overs, runs, wickets, economy, bowling_avg, bowling_sr, dot_balls, dot_balls_pct, fours, sixes, bbf)
```

```python
agg_player_bowling_table = pd.DataFrame()
agg_player_bowling_table["player_id"] = player_id_col
agg_player_bowling_table["season_id"] = season_id_col
agg_player_bowling_table["matches"] = matches_col
agg_player_bowling_table["innings"] = innings_col
agg_player_bowling_table["balls_bowled"] = balls_bowled_col
agg_player_bowling_table["overs"] = overs_col
agg_player_bowling_table["runs_conceded"] = runs_conceded_col
agg_player_bowling_table["wickets"] = wickets_col
agg_player_bowling_table["economy"] = economy_col
agg_player_bowling_table["bowling_average"] = bowling_avg_col
agg_player_bowling_table["bowling_strike_rate"] = bowling_sr_col
agg_player_bowling_table["dot_balls"] = dot_balls_col
agg_player_bowling_table["dot_balls_pct"] = dot_balls_pct_col
agg_player_bowling_table["fours_conceded"] = fours_conceded_col
agg_player_bowling_table["sixes_conceded"] = sixes_conceded_col
agg_player_bowling_table["best_bowling_figures"] = best_bowling_figs_col

```

```python
team_id_col = []
season_id_col = []
matches_col = []
wins_col = []
losses_col = []
ties_col = []
no_results_col = []
win_pct_col = []
runs_scored_col = []
runs_conceded_col = []
wickets_taken_col = []
wickets_lost_col = []
avg_runs_scored_col = []
avg_runs_conceded_col = []
chasing_matches_col = []
chasing_wins_col = []
chasing_win_pct_col = []
defending_matches_col = []
defending_wins_col = []
defending_win_pct_col = []
fours_col = []
sixes_col = []


season_map = season_dim_table.set_index("season")["season_id"]
team_map = team_dim_table.set_index("team_name")["team_id"]

for team in team_dim_table["team_name"].unique():
    for season in season_dim_table["season"].unique():

        values = cleaned_ipl_data[(cleaned_ipl_data["season"]==season) & ((cleaned_ipl_data["batting_team"]==team) | (cleaned_ipl_data["bowling_team"]==team))]

        if values.shape[0]==0:
            continue
        
        # -------------------------------------------------------------
        #                       Calculate Values
        # -------------------------------------------------------------
        matches = values["match_id"].nunique()
        try:
            wins = values[["match_id","match_won_by"]].drop_duplicates()["match_won_by"].value_counts()[team]
        except KeyError:
            wins = 0

        try:
            ties = values[["match_id","result_type"]].drop_duplicates()["result_type"].value_counts()["TIE"]
        except KeyError:
            ties = 0

        try:
            no_results = values[["match_id","result_type"]].drop_duplicates()["result_type"].value_counts()["NO RESULT"]
        except KeyError:
            no_results = 0

        losses = matches - wins - ties - no_results
        # print(wins, matches)
        win_pct = round(wins/matches*100, 2)
        runs_scored = values[values["batting_team"]==team]["runs_total"].sum()
        runs_conceded = values[values["batting_team"]!=team]["runs_total"].sum()
        wicks_taken = values[values["batting_team"]!=team]["wicket_kind"].value_counts().sum()
        wicks_lost = values[values["batting_team"]==team]["wicket_kind"].value_counts().sum()
        avg_runs_scored = round(runs_scored/matches, 2)
        avg_runs_conceded = round(runs_conceded/matches, 2)


        chasing_matches = values[(values["innings"]%2==0) & (values["batting_team"]==team) ]["match_id"].nunique()
        chasing_ties_no_results = values[(values["innings"]%2==0) & (values["batting_team"]==team)  & (values["result_type"].isin(["TIE", "NO RESULT"]))]["match_id"].nunique()

        try:
            chasing_wins = values[(values["innings"]%2==0) & (values["batting_team"]==team)][["match_id","match_won_by"]].drop_duplicates()["match_won_by"].value_counts()[team]
        except KeyError:
            chasing_wins = 0
        
        chasing_win_pct = round((chasing_wins/(chasing_matches-chasing_ties_no_results))*100, 2)

        defending_matches = values[(values["innings"]%2!=0) & (values["batting_team"]==team) ]["match_id"].nunique()
        defending_ties_no_results = values[(values["innings"]%2!=0) & (values["batting_team"]==team)  & (values["result_type"].isin(["TIE", "NO RESULT"]))]["match_id"].nunique()

        try:
            defending_wins = values[(values["innings"]%2!=0) & (values["batting_team"]==team)][["match_id","match_won_by"]].drop_duplicates()["match_won_by"].value_counts()[team]
        except KeyError:
            defending_wins = 0

        defending_win_pct = round((defending_wins/(defending_matches-defending_ties_no_results))*100, 2)

        try:
            fours = values["runs_batter"].value_counts()[4]
        except:
            fours = 0

        try:
            sixes = values["runs_batter"].value_counts()[6]
        except:
            sixes = 0
        
        # -------------------------------------------------------------
        #                       Append Values
        # -------------------------------------------------------------
        team_id_col.append(team_map[team])
        season_id_col.append(season_map[season])
        matches_col.append(matches)
        wins_col.append(wins)
        losses_col.append(losses)
        ties_col.append(ties)
        no_results_col.append(no_results)
        win_pct_col.append(win_pct)
        runs_scored_col.append(runs_scored)
        runs_conceded_col.append(runs_conceded)
        wickets_taken_col.append(wicks_taken)
        wickets_lost_col.append(wicks_lost)
        avg_runs_scored_col.append(avg_runs_scored)
        avg_runs_conceded_col.append(avg_runs_conceded)
        chasing_matches_col.append(chasing_matches)
        chasing_wins_col.append(chasing_wins)
        chasing_win_pct_col.append(chasing_win_pct)
        defending_matches_col.append(defending_matches)
        defending_wins_col.append(defending_wins)
        defending_win_pct_col.append(defending_win_pct)
        fours_col.append(fours)
        sixes_col.append(sixes)

    
```

### Agg Venue Table

```python
venue_id_col = []
season_id_col = []
matches_col = []
avg_first_innings_score_col = []
avg_second_innings_score_col = []
highest_total_col = []
lowest_total_col = []
highest_total_chased_col = []
batting_first_wins_col = []
batting_second_wins_col = []
batting_first_win_pct_col = []
batting_second_win_pct_col = []
avg_wickets_col = []
fours_col = []
sixes_col = []

season_map = season_dim_table.set_index("season")["season_id"]
venue_map = venue_dim_table.set_index("venue")["venue_id"]

for stadium in venue_dim_table["venue"].unique():
    for season in season_dim_table["season"].unique():

        
        values = cleaned_ipl_data[(cleaned_ipl_data["venue"]==stadium) & (cleaned_ipl_data["season"]==season)]

        if values.empty:
            continue


        matches = values["match_id"].nunique()

        values = values[(~values["result_type"].isin(["TIE", "NO RESULT"]))]

        first_innings_scores = values[values["innings"]==1].groupby(["match_id"])["runs_total"].sum().values.tolist()
        second_innings_scores = values[values["innings"]==2].groupby(["match_id"])["runs_total"].sum().values.tolist()

        avg_first_innings_score = round(np.mean(first_innings_scores),2)
        avg_second_innings_score = round(np.mean(second_innings_scores),2)

        successful_chases = [
            second
            for first, second in zip(first_innings_scores, second_innings_scores)
            if second > first
        ]
        highest_total_chased = max(successful_chases) if successful_chases else None

        batting_first_wins = sum(
            first > second
            for first, second in zip(first_innings_scores, second_innings_scores)
        )

        batting_second_wins = sum(
            second > first
            for first, second in zip(first_innings_scores, second_innings_scores)
        )

        batting_first_win_pct = round((batting_first_wins/matches)*100,2)
        batting_second_win_pct = round((batting_second_wins/matches)*100,2)

        highest_total = max(first_innings_scores+second_innings_scores)
        lowest_total = min(first_innings_scores+second_innings_scores)

        wickets = values.groupby(["match_id", "batting_team"])["wicket_kind"].count().tolist()
        avg_wickets = round(np.mean(wickets))

        try:
            fours = values["runs_batter"].value_counts()[4]
        except:
            fours = 0

        try:
            sixes = values["runs_batter"].value_counts()[6]
        except:
            sixes = 0
        
        venue_id_col.append(venue_map[stadium])
        season_id_col.append(season_map[season])
        matches_col.append(matches)
        avg_first_innings_score_col.append(avg_first_innings_score)
        avg_second_innings_score_col.append(avg_second_innings_score)
        highest_total_col.append(highest_total)
        lowest_total_col.append(lowest_total)
        highest_total_chased_col.append(highest_total_chased)
        batting_first_wins_col.append(batting_first_wins)
        batting_second_wins_col.append(batting_second_wins)
        batting_first_win_pct_col.append(batting_first_win_pct)
        batting_second_win_pct_col.append(batting_second_win_pct)
        avg_wickets_col.append(avg_wickets)
        fours_col.append(fours)
        sixes_col.append(sixes)
```

```python
player_id_col = []
season_id_col = []
phase_col = []
matches_col = []
innings_col = []
runs_col = []
balls_col = []
average_col = []
strike_rate_col = []
highest_score_col = []
fours_col = []
sixes_col = []
dismissals_col = []
dot_ball_pct_col = []
boundary_pct_col = []

for season_id,vals in match_dim_table.groupby(["season_id"]):
    
    match_ids = vals['match_id']
    season_id = season_id[0]
    print(f" Found {len(match_ids)} matches for IPL season {season_id}")

    for key,values in enriched_delivery_fact_table[enriched_delivery_fact_table["match_id"].isin(match_ids)].groupby(["batter_id","phase"]):
    
    
        player_id, phase = key

        matches = values["match_id"].nunique()
        innings = values[["match_id","innings"]].drop_duplicates().shape[0]
        match_runs = values.groupby(["match_id"])["runs_batter"].sum().tolist()
        runs = sum(match_runs)
        balls = values["batter_balls"].shape[0]
        average = round(runs/matches, 2)
        strike_rate = round(runs/balls*100, 2)
        highest_score = max(match_runs)
        dismissals = values["wicket_kind"].value_counts().sum()
        
        try:
            fours = values["runs_batter"].value_counts()[4]
        except:
            fours = 0

        try:
            sixes = values["runs_batter"].value_counts()[6]
        except:
            sixes = 0
        
        try:
            dot_balls_pct = round(values["runs_total"].value_counts()[0]/balls * 100, 2)
        except KeyError:
            dot_balls_pct = 0
        
        boundary_pct = round((fours+sixes)/balls*100, 2)

        player_id_col.append(player_id)
        season_id_col.append(season_id)
        phase_col.append(phase)
        matches_col.append(matches)
        innings_col.append(innings)
        runs_col.append(runs)
        balls_col.append(balls)
        average_col.append(average)
        strike_rate_col.append(strike_rate)
        highest_score_col.append(highest_score)
        fours_col.append(fours)
        sixes_col.append(sixes)
        dismissals_col.append(dismissals)
        dot_ball_pct_col.append(dot_balls_pct)
        boundary_pct_col.append(boundary_pct)
```

### Agg Bowler Phase Table

```python
player_id_col = []
season_id_col = []
phase_col = []
matches_col = []
innings_col = []
balls_bowled_col = []
overs_col = []
runs_conceded_col = []
wickets_col = []
economy_col = []
bowling_avg_col = []
bowling_sr_col = []
dot_balls_col = []
dot_ball_pct_col = []
fours_conceded_col = []
sixes_conceded_col = []

for season_id,vals in match_dim_table.groupby(["season_id"]):
    
    match_ids = vals['match_id']
    season_id = season_id[0]
    print(f" Found {len(match_ids)} matches for IPL season {season_id}")

    for key,values in enriched_delivery_fact_table[enriched_delivery_fact_table["match_id"].isin(match_ids)].groupby(["bowler_id","phase"]):
    
    
        player_id, phase = key

        matches = values["match_id"].nunique()
        innings = values[["match_id","innings"]].drop_duplicates().shape[0]
        balls_bowled = values["batter_balls"].shape[0]
        overs = np.floor(balls_bowled/6)
        match_runs = values.groupby(["match_id"])["runs_bowler"].sum().tolist()
        runs_conceded = sum(match_runs)
        try:
            wickets = values["is_bowler_wicket"].value_counts()[True]
        except:
            wickets = 0
        
        economy = round(runs_conceded/overs, 2)
        
        try:
            bowling_avg = round(runs_conceded/wickets, 2)
        except ZeroDivisionError:
            bowling_avg = 0
        try:
            legal_balls_bowled = values["valid_ball"].value_counts()[True]
        except KeyError:
            legal_balls_bowled = 0
        try:
            bowling_strike_rate = round((legal_balls_bowled/wickets), 2)
        except ZeroDivisionError:
            bowling_strike_rate = 0
        
        try:
            fours_conceded = values["runs_batter"].value_counts()[4]
        except:
            fours_conceded = 0

        try:
            sixes_conceded = values["runs_batter"].value_counts()[6]
        except:
            sixes_conceded = 0
        
        try:
            dot_balls = values["runs_total"].value_counts()[0]
            dot_balls_pct = round(dot_balls/balls_bowled * 100, 2)
        except KeyError:
            dot_balls = 0
            dot_balls_pct = 0
        

        player_id_col.append(player_id)
        season_id_col.append(season_id)
        phase_col.append(phase)
        matches_col.append(matches)
        innings_col.append(innings)
        balls_bowled_col.append(balls_bowled)
        overs_col.append(overs)
        runs_conceded_col.append(runs_conceded)
        wickets_col.append(wickets)
        economy_col.append(economy)
        bowling_avg_col.append(bowling_avg)
        bowling_sr_col.append(bowling_strike_rate)
        dot_balls_col.append(dot_balls)
        dot_ball_pct_col.append(dot_balls_pct)
        fours_conceded_col.append(fours_conceded)
        sixes_conceded_col.append(sixes_conceded)
```

```python
agg_bowler_vs_phase = pd.DataFrame()

agg_bowler_vs_phase["player_id"] = player_id_col
agg_bowler_vs_phase["season_id"] = season_id_col
agg_bowler_vs_phase["phase"] = phase_col
agg_bowler_vs_phase["matches"] = matches_col
agg_bowler_vs_phase["innings"] = innings_col
agg_bowler_vs_phase["balls_bowled"] = balls_bowled_col
agg_bowler_vs_phase["overs"] = overs_col
agg_bowler_vs_phase["runs_conceded"] = runs_conceded_col
agg_bowler_vs_phase["wickets"] = wickets_col
agg_bowler_vs_phase["economy"] = economy_col
agg_bowler_vs_phase["bowling_avg"] = bowling_avg_col
agg_bowler_vs_phase["bowling_strike_rate"] = bowling_sr_col
agg_bowler_vs_phase["dot_balls"] = dot_balls_col
agg_bowler_vs_phase["dot_balls_pct"] = dot_ball_pct_col
agg_bowler_vs_phase["fours_conceded"] = fours_conceded_col
agg_bowler_vs_phase["sixes_conceded"] = sixes_conceded_col
```
